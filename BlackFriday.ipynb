```python
"""
Data representation of black friday sale from a shop
Shown that male customers aged 26-35 are most active and bring highest revenue.

data URL is https://www.kaggle.com/mehdidag/black-friday
"""

%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('BlackFriday.csv')
df.fillna(0, inplace=True)
age_order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']

# distribution of total purchase
plt.figure()
plt.hist(df['Purchase'], bins=50)
plt.xlabel('Purchase')
plt.ylabel('Count')

# purchase count on users, products, gender and age groups separately
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
sns.countplot('User_ID', data=df)
plt.xlabel('User ID')
plt.ylabel('')
plt.title('Total count')
plt.xticks([])

plt.subplot(222)
sns.countplot('Product_ID', data=df)
plt.xlabel('Product ID')
plt.ylabel('')
plt.title('Total count')
plt.xticks([])

plt.subplot(223)
sns.countplot('Gender', data=df)
plt.xlabel('Gender')
plt.ylabel('')
plt.title('Total count')
plt.xticks(np.arange(2), ('Female', 'Male'))

plt.subplot(224)
sns.countplot('Age', order=age_order, data=df)
plt.xlabel('Age')
plt.ylabel('')
plt.title('Total count')
plt.xticks(rotation=70)

# total purchase on age groups and gender
fig = plt.figure(figsize=(10,7))
total_purchase_age = df.groupby(['Age', 'Gender']).aggregate({'Purchase':np.sum}).reset_index()
sns.barplot(x='Age', y='Purchase', hue='Gender', data=total_purchase_age)
plt.xlabel('Age')
plt.ylabel('')
plt.title('Total purchase')
plt.xticks(rotation=70)

plt.show()
```
