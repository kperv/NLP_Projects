```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# dataURL is https://www.kaggle.com/mehdidag/black-friday
df = pd.read_csv('BlackFriday.csv')

ratio = 100 * df.isnull().sum() / df['User_ID'].count()
ratio = ratio.round(2)
ratio[ratio > 0]

df.fillna(0, inplace=True)
df.nunique()


plt.figure()
plt.hist(df['Purchase'], bins=50)
plt.xlabel('Purchase')
plt.ylabel('Count')

plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
sns.countplot('User_ID', data=df)
plt.xlabel('User ID')
plt.ylabel('')
plt.title('Total count')
plt.xticks([])

plt.subplot(222)
sns.countplot('Product_ID', data=df)
plt.xlabel('Product ID')
plt.ylabel('')
plt.title('Total count')
plt.xticks([])

plt.subplot(223)
sns.countplot('Gender', data=df)
plt.xlabel('Gender')
plt.ylabel('')
plt.title('Total count')
plt.xticks([])

age_order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']

plt.subplot(224)
sns.countplot('Age', order=age_order, data=df)
plt.xlabel('Age')
plt.ylabel('')
plt.title('Total count')
plt.xticks(rotation=70)

plt.show()
```

```python
age_order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']

fig = plt.figure(figsize=(15,10))
plt.subplot(121)
sns.countplot('Age', order=age_order, hue='Gender', data=df)
plt.xlabel('Age')
plt.ylabel('')
plt.title('Total count')
plt.xticks(rotation=70)
plt.title('Number of customers')
plt.legend(['Female', 'Male'], frameon=True)


total_purchase_age = df.groupby(['Age', 'Gender']).aggregate({'Purchase':np.sum}).reset_index()
plt.subplot(122)
sns.barplot(x='Age', y='Purchase', hue='Gender', data=total_purchase_age)
plt.xlabel('Age')
plt.ylabel('')
plt.title('Total purchase')
plt.xticks(rotation=70)
plt.legend().set_visible(False)

plt.show()
```
